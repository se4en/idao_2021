```python
import os
import numpy as np
import pandas as pd
import cv2
from PIL import Image
from tqdm import tqdm

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
ER_path = "idao_dataset/train_prep/ER/"
NR_path = "idao_dataset/train_prep/NR/"
```

```python
train_ER = os.listdir(ER_path)
train_NR = os.listdir(NR_path)
print(f"ER={len(train_ER)} NR={len(train_NR)}")
```

```python
def flip_rotate(path, draw=False):
    '''
    Создает 7 новых фото - 3 поворота по 90* и 4 отражения
    '''
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    img_1 = np.fliplr(img)
    img_2 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    img_3 = np.flipud(img_2)
    img_4 = cv2.rotate(img, cv2.ROTATE_180)
    img_5 = np.fliplr(img_4)
    img_6 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    img_7 = np.flipud(img_6)
    
    if draw:
        fig = plt.figure()
        fig.set_size_inches(20,10)
        plt.subplot(241),plt.imshow(img, aspect='auto')
        plt.subplot(241).title.set_text("Orig")
        plt.subplot(242),plt.imshow(img_1, aspect='auto')
        plt.subplot(242).title.set_text("-1(+flip)")
        plt.subplot(243),plt.imshow(img_2, aspect='auto')
        plt.subplot(243).title.set_text("-2(+90)")
        plt.subplot(244),plt.imshow(img_3, aspect='auto')
        plt.subplot(244).title.set_text("-3(+90+flip)")
        plt.subplot(245),plt.imshow(img_4, aspect='auto')
        plt.subplot(245).title.set_text("-4(+180)")
        plt.subplot(246),plt.imshow(img_5, aspect='auto')
        plt.subplot(246).title.set_text("-5(+180+flip)")
        plt.subplot(247),plt.imshow(img_6, aspect='auto')
        plt.subplot(247).title.set_text("-6(+270)")
        plt.subplot(248),plt.imshow(img_7, aspect='auto')
        plt.subplot(248).title.set_text("-7(+270+flip)")
        plt.show()
    return [img, img_1, img_2, img_3, img_4, img_5, img_6, img_7]
```

```python
_ = flip_rotate(ER_path + train_ER[2], True)
```

#### Flip ER

```python
new_folder_ER = "idao_dataset/train_prep_max/ER/"
```

```python
for i in tqdm(range(len(train_ER))):
    name = train_ER[i]
    imgs = flip_rotate(ER_path + name)
    for j, img in enumerate(imgs):
        cv2.imwrite(new_folder_ER + name[:-4] + f"-{j}.png", img) 
```

#### Flip NR

```python
new_folder_NR = "idao_dataset/train_prep_max/NR/"
```

```python
for i in tqdm(range(len(train_NR))):
    name = train_NR[i]
    imgs = flip_rotate(NR_path + name)
    for j, img in enumerate(imgs):
        cv2.imwrite(new_folder_NR + name[:-4] + f"-{j}.png", img) 
```
